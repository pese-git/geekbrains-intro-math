## 1 Исследовать ряд на сходимость, используя признак д’Аламбера:

$$
\sum_{n=1}^{\infty}\frac{n^n}{(!n)^2}
$$

### Решение



Признак Даламбера: Рассмотрим положительный числовой ряд $$\sum_{n=1}^{\infty}a^n$$. Если существует 
предел отношения последующего члена к предыдущему: $\lim_{n \to \infty}\frac{a_{n+1}}{a_n} = D$ , то:


- a) При $D < 1$ ряд сходится. В частности, ряд сходится при .
- б) При $D > 1$  ряд расходится. В частности, ряд расходится при .
- в) При $D = 1$  признак не дает ответа.


$$
\sum_{n=1}^{\infty}\frac{n^n}{(!n)^2} = \frac{1^1}{1^2} + \frac{2^2}{(1 * 2)^2} + \frac{3^3}{(1 * 2 *3)^2} + ... + \frac{n^n}{(!n)^2} + \frac{(n+1)^{n+1}}{(!(n+1))^2}
$$

$$
\lim_{n \to \infty}\frac{\frac{(n+1)^{n+1}}{(!(n+1))^2}}{\frac{n^n}{(!n)^2}} = \lim_{n \to \infty} \frac{(n+1)^{n+1}}{(!(n+1))^2} * \frac{(!n)^2}{n^n} = \lim_{n \to \infty} \frac{(n+1)^{n}(n+1)}{(!(n+1))^2} * \frac{(!n)^2}{n^n}
$$


## 2 Исследовать ряд на сходимость, используя радикальный признак Коши:

$$
\sum_{n=1}^{\infty}\frac{n}{2^n}
$$

### Решение


Радикальный признак Коши: Рассмотрим положительный числовой ряд $$\sum_{n=1}^{\infty}a^n$$. Если существует предел: , то $\lim_{n \to \infty}\sqrt[n]{a_n}$:

- a) При $D < 1$ ряд сходится. В частности, ряд сходится при .
- б) При $D > 1$  ряд расходится. В частности, ряд расходится при .
- в) При $D = 1$  признак не дает ответа.



$$
\sum_{n=1}^{\infty}\frac{n}{2^n} => D = \lim_{n \to \infty} \sqrt[n]{\frac{n}{2^n}} = \lim_{n \to \infty} (\frac{n}{2^n})^{\frac{1}{n}}
$$

В данном случае степень $\frac{1}{n} \to 0$ стремится к 0, и это значит что все выражение $\lim_{n \to \infty} (\frac{n}{2^n})^{\frac{1}{n}}$ будет стремится к 1

$$
D = \lim_{n \to \infty} \sqrt[n]{\frac{n}{2^n}} = \lim_{n \to \infty} (\frac{n}{2^n})^{\frac{1}{n}} = 1
$$


И по условию при $D=1$ признак не дает ответа.

## 3 Исследовать ряд на сходимость, используя признак Лейбница:


$$
\sum_{n=1}^{\infty}\frac{(-1)^n}{n + \ln{n}}
$$

### Решение


Признак Лейбница: Если члены знакочередующегося ряда монотонно убывают по модулю, то ряд сходится $$\lim_{n \to \infty}|a_n| = 0$$.

Или в два пункта:

- 1) Ряд является знакочередующимся.
- 2) Члены ряда убывают по модулю: , причём, убывают монотонно.



$$
\sum_{n=1}^{\infty}\frac{(-1)^n}{n + \ln{n}}
$$


1) Проверка ряда на знакочередование.

$$
\sum_{n=1}^{\infty}\frac{(-1)^n}{n + \ln{n}} = - \frac{1}{1 + \ln{1}} + \frac{1}{2 + \ln{2}} - \frac{1}{3 + \ln{3}}
$$

Ряд является знакочередуешимся

2) Убывают ли члены ряда по модулю?

$$
\lim_{n \to \infty} |\frac{(-1)^n}{n + \ln{n}}| = \lim_{n \to \infty} |\frac{1}{n + \ln{n}}| = \frac{1}{\infty} = 0
$$


члены ряда убывают по модулю, и из этого автоматически следует его сходимость
