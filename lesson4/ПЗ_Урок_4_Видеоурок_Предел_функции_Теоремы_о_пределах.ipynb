## 4 Найти предел:

a. $\lim_{x\to 0} \frac{3x^3 -2x^2}{4x^2}  $

b.* $\lim_{x\to 0} \frac{\sqrt{1+x} - 1}{\sqrt[3]{1+x} - 1} $

c.* $\lim_{x\to \infty} (\frac{x + 3}{x})^{4x + 1} $

### Решение:

a. $\lim_{x\to 0} \frac{3x^3 -2x^2}{4x^2} = (\frac{0}{0}) = \frac{x^2(3x -2)}{4x^2} = \frac{(3x - 2)}{4} = \frac{0 - 2}{4} = - \frac{2}{4} = - \frac{1}{2}$

b.* $\lim_{x\to 0} \frac{\sqrt{1+x} - 1}{\sqrt[3]{1+x} - 1} = (\frac{0}{0}) = \frac{\sqrt{1+x} - 1}{\sqrt[3]{1+x} - 1} * \frac{\sqrt{1+x} + 1}{\sqrt{1+x} + 1} = \frac{1 + x -1}{(\sqrt[3]{1+x} - 1)(\sqrt{1+x} + 1)} = \frac{x}{(\sqrt[3]{1+x} - 1)(\sqrt{1+x} + 1)}$

# Тема “Теоремы о пределах”

## 1. Найти предел:

a. $\lim_{x\to 0} \frac{sin(2x)}{4x}  $

b. $\lim_{x\to 0} \frac{x}{sin(x)}  $

c. $\lim_{x\to 0} \frac{x}{arcsin(x)}  $

d. $\lim_{x\to \infty} (\frac{4x + 3}{4x - 3})^{6x}  $

e. $\lim_{x\to \infty} \frac{\sin{x} + \ln{x}}{x}  $

f. $\lim_{x\to 0} \frac{\sin(x) + \ln(x)}{x}  $


### Решение:

a. $\lim_{x\to 0} \frac{sin(2x)}{4x} = (\frac{0}{0}) =  \frac{sin(2x)}{2 * 2x} = \frac{1}{2} * \frac{sin(2x)}{2x} = \frac{1}{2} $, где $ \lim_{x\to 0} \frac{sin(2x)}{2x} = 1 $ - первый замечательный предел

b. $\lim_{x\to 0} \frac{x}{sin(x)} = (\frac{0}{0}) = 1 $, где $\lim_{x\to 0} \frac{x}{sin(x)} = 1 $ - первый замечательный предел

c. $\lim_{x\to 0} \frac{x}{arcsin(x)} = (\frac{0}{0}) = 1 $, arcsin является следствием первого замечательного предела вида $\lim_{x \to 0} \frac{x}{sin(x)}$
